```python
import pandas as pd
import mysql.connector

# Read the CSV file into a DataFrame
df = wilayah

# Connect to the MySQL database
conn = mysql.connector.connect(
    host='localhost',
    user='root',
    password='12345',
    database='bpjs'
)

# Create a cursor object to execute SQL statements
cursor = conn.cursor()

# Create a table with the appropriate schema
table_name = 'wilayah'
create_table_query = '''
    CREATE TABLE {} (
        NoLink int not null, 
        Provinsi varchar(50), 
        KotaKab varchar(100), 
        Link varchar(1000), 
        TipeFaskes varchar(50)
    )
'''.format(table_name)
cursor.execute(create_table_query)

# Insert the DataFrame data into the table
insert_query = 'INSERT INTO {} (NoLink, Provinsi, KotaKab, Link, TipeFaskes) VALUES (%s, %s, %s, %s, %s)'.format(table_name)
for row in df.itertuples(index=False):
    cursor.execute(insert_query, row)

# Commit the changes and close the cursor and connection
conn.commit()
cursor.close()
conn.close()
```

```python
import pandas as pd
import mysql.connector

# Read the CSV file into a DataFrame
df = faskes

# Replace '-' values with None
df.replace('-', 1, inplace=True)

# Connect to the MySQL database
conn = mysql.connector.connect(
    host='localhost',
    user='root',
    password='12345',
    database='bpjs'
)

# Create a cursor object to execute SQL statements
cursor = conn.cursor()

# Create a table with the appropriate schema
table_name = 'faskes'
create_table_query = '''
    CREATE TABLE {} (
        No int not null,
        KodeFaskes varchar(20),
        NamaFaskes varchar(100), 
        LatLongFaskes varchar(150),
        AlamatFaskes varchar(100),
        TelpFaskes varchar(70)
    )
'''.format(table_name)
cursor.execute(create_table_query)

# Insert the DataFrame data into the table
insert_query = 'INSERT INTO {} (No, KodeFaskes, NamaFaskes, LatLongFaskes, AlamatFaskes, TelpFaskes) VALUES (%s, %s, %s, %s, %s, %s)'.format(table_name)
for row in df.itertuples(index=False):
    cursor.execute(insert_query, row)

# Commit the changes and close the cursor and connection
conn.commit()
cursor.close()
conn.close()
```

### Insight 
1. Jika andi sakit gigi dan tinggal di provinsi NAD dimana tempat yang sesuai untuk andi berobat ? 

```python
%%sql 
SELECT KotaKab, Link
FROM wilayah
WHERE Provinsi = 'Nanggroe Aceh Darussalam' AND TipeFaskes = 'Dokter Gigi'
```

2. Jika ada seorang mengalami kecelakaan da harus dibawa ke RSU Dimanakah dia harus dibawa ? 

```python
%%sql 
SELECT No, KodeFaskes, NamaFaskes, LatLongFaskes, AlamatFaskes, TelpFaskes
FROM Faskes
WHERE NamaFaskes like 'RSU%'
```

## Grant Access to person

Mebuat user_1, memberikannya seluruh akses, menghapus haknya untuk update sehingga user_1 hanya tidak bisa update 

```python
%%sql
create user 'user1'@'localhost'
```

```python
%%sql
grant all on bpjs.* to 'user1'@'localhost'
```

```python
%%sql
revoke update on bpjs.* from 'user1'@'localhost';
```

```python
%%sql
show grants for 'user1'@'localhost'
```
