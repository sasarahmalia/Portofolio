```python
data = pandas.read_csv('/content/drive/MyDrive/TUBES/package_tourism.csv')
data.drop(["Package","Place_Tourism4", "Place_Tourism5"], axis = 1, inplace=True)
data
```

```python
class Stack:
  def __init__(self, data):
    self.data = data
    self.kolom = []
    for i in self.data:
      self.kolom.append(i)
#melihat jumlah kolom dan baris
  def __len__(self):
    return f"{len(list(self.data[self.kolom[0]]))} rows Ã— {len(self.kolom)} columns"
#mengecek data apakah data kosong
  def isempty(self):
    return len(list(self.data[self.kolom[0]])) == 0
#nampilin data
  def view(self):
    return self.data
#menambah data dari index terakhir ke data frame
#dataframe diubah ke list kemudian ditambah lalu diubah kembali ke dataframe
  def push(self, item):
    kol = {}
    j = 0
    for i in self.kolom:
      kol[j] = list(self.data[i])
      j += 1
    k = 0
    data = {}
    for j in item:
      kol[k].append(j)
      k += 1
    p = 0
    for m in self.kolom:
      data[m] = kol[p]
      p += 1
    self.data = pandas.DataFrame(data)
#menghapus dan menampilkan data trakhir
  def pop(self):
    if self.isempty():
      raise Exception('Stack kosong')
    else:
      data = {}
      df_a = self.data.tail(1)
      for i in self.kolom:
        data[i] = list(self.data[i])
        data[i].pop()
      self.data = pandas.DataFrame(data)
      return df_a
#menampilkan data paling terakhir
  def top(self):
    if self.isempty():
      raise Exception('Stack kosong')
    return self.data.tail(1)
```

```python
wisata = Stack(data)
wisata.__len__()
wisata.isempty()
wisata.view()
wisata.push(["Lampung","Pantai Mutun " , "Gunung Sagi", "Puncak Mas"])
wisata.view()
```
